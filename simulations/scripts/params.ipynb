```python
from model_wrappers import *
import sys
sys.path.append('../')
# from simulate_mofa import simulate_data
import itertools
import pandas as pd
```

## sim_par_1

### baseline

```python
N_experiments = 20
N = 100
K = 5 
noise_std = 1
sparsity = 0.75 
M = 3
D = [400 for m in range(M)] 
```

```python
params_grid = {'N': [N],
                 'K': [K],
                 'D': [D],
                 'noise_std': [noise_std],
                 'sparsity': [sparsity],
                 'seed': np.arange(0, N_experiments, 1).tolist()}
keys = params_grid.keys()
params_grid = list(itertools.product(*params_grid.values()))
df_baseline = pd.DataFrame(params_grid, columns=keys)
```

### x-axis: K, noise, sparsity

```python
params_grid_K = {'N': [N],
                 'K': [1, 7, 10, 15, 20, 25],
                 'D': [D],
                 'noise_std': [noise_std],
                 'sparsity': [sparsity],
                 'seed': np.arange(0, N_experiments, 1).tolist()}
keys = params_grid_K.keys()
params_grid_K = list(itertools.product(*params_grid_K.values()))
df_K = pd.DataFrame(params_grid_K, columns=keys)
```

```python
params_grid_noise = {'N': [N],
                 'K': [K],
                 'D': [D],
                 'noise_std': [0.25, 0.5, 0.75, 1.25, 1.5, 1.75, 2.0, 2.5], 
                 'sparsity': [sparsity],
                 'seed': np.arange(0, N_experiments, 1).tolist()}
keys = params_grid_noise.keys()
params_grid_noise = list(itertools.product(*params_grid_noise.values()))
df_noise = pd.DataFrame(params_grid_noise, columns=keys)
```

```python
params_grid_sparsity = {'N': [N],
                 'K': [K],
                 'D': [D],
                 'noise_std': [noise_std],
                 'sparsity': [0.65, 0.70, 0.8, 0.85],
                 'seed': np.arange(0, N_experiments, 1).tolist()}
keys = params_grid_sparsity.keys()
params_grid_sparsity = list(itertools.product(*params_grid_sparsity.values()))
df_sparsity = pd.DataFrame(params_grid_sparsity, columns=keys)
```

### save params

```python
df = pd.concat([df_baseline, df_K, df_noise, df_sparsity], ignore_index=True)
```

```python
df.to_csv('param_sim_1.csv')
```
